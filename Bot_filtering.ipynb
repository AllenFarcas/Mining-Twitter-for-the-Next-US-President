```python
import ast
import json
from json_excel_converter import Converter 
from json_excel_converter.xlsx import Writer
import pandas as pd
import csv
import numpy as np


def create_botlist(input_file_path,tweet_limit):
    
            id_data_vec=[]
            date_vec=[]
            pd.set_option("display.max_rows", None, "display.max_columns", None)
            pd.options.display.max_seq_items = None
            count = 0
            line_list = []

            cnt=0
            bot_list=[]
            dfcheck = pd.DataFrame([1], columns=['Sep 15'],index=[1])

            #for mon in ['Aug','Sep','Oct','Nov']:
            #    for dat in range(1,32):
            cnt=0
            with open(input_file_path) as json_file: 

                                for line in json_file:
                                        cnt=cnt+1

                                        data=(json.loads(line))


                                        date=data['created_at'][4:10]
                                        
                                        id_data=int(data['u_id_str'])
                                        

                                        dfcheck=dfcheck.fillna(0)
                                        
                                        if (id_data in id_data_vec) and (date in date_vec):
                                                   
                                                vec_check=dfcheck.at[id_data, date]
                                                #print(vec_check.type)
                                                if (dfcheck.at[id_data, date].size)==1:
                                                    dfcheck.at[id_data, date]=vec_check+1
                                                else:
                                                    dfcheck.at[id_data, date]=vec_check[0]+1
                                                        
                                            
                                                   
                                                print(dfcheck.at[id_data, date])
                                                if (np.max(dfcheck.at[id_data, date])==tweet_limit):
                                                            bot_list.append(id_data)
                                                   
                                                
                                        else:

                                                df2 = pd.DataFrame(1,index=[id_data],columns=[date])

                                                dfcheck=dfcheck.append(df2)
                                                
                                        id_data_vec.append(id_data)
                                        date_vec.append(date)
                                        
                                                
                                        
                                        
            return bot_list
                                    

#dfcheck=dfcheck.sort_index()
#print(dfcheck)    


def bot_remove(input_file,output_file,bot_list):
            with open(input_file) as json_file: 

                                for line in json_file:


                                        data=(json.loads(line))


                                        date=data['created_at'][4:10]
                                        id_data=data['user']['screen_name']

                                        if not(id_data in bot_list):
                                            with open(output_file, 'a') as outfile:
                                                    json.dump(data, outfile)

if __name__=="__main__": 
    bot_list=create_botlist('dataset_filt.jsonl',150)
    
    print(bot_list)
                            
    bot_remove('dataset_filt.jsonl','bot_filtered.jsonl',bot_list)
    

```
